```python
#Normalize Images to Extract Prototypes
# load image
def preproc_im(imname):
    image = Image.open(imname)
    pixels = np.asarray(image)
    pixels = pixels.astype('float32')
    mean, std = pixels.mean(), pixels.std()
    outim = (pixels - mean) / std
    return(outim)
# for rawim in glob('data/RedLights2011_Medium/*.jpg'):
#     outim=preproc_im(rawim)
#     outim.save('data/RedLights_Normed/'+str.split(rawim,'\\')[-1])
```

```python
def length(a1):
    lenout = np.sqrt(np.sum(a1**2, axis=-1))
    return(lenout)
def angle(a1, a2, indenom=False):
    if not indenom:
        denom = (length(a1) * length(a2))
    else:
        denom=indenom
    numer = np.dot(a1, a2)
    arccosin = numer/denom
    angrads = np.arccos(arccosin)
    return(angrads)
```

```python
testim=preproc_im('data/RedLights2011_Medium/RL-190.jpg')[:,:,:3]
print(np.min(testim),np.max(testim))
testim=np.divide(testim,(np.expand_dims(np.sqrt(np.sum(testim**2,axis=-1)),-1)))
detmvals=[]
detbbs=[]
kerns=glob('data/kernels/*.png')
# chunksize,chunkrem=len(kerns)//10,len(kerns)%10
# chunks=[]
# for chunknum in range(10):
#     chunklist=kerns[chunknum*chunksize:chunknum*chunksize+chunksize]
#     chunks.appemnd(chunklist)

for kernname in kerns:
    kern=preproc_im(kernname)[:,:,:3]
    kern=np.divide(kern,(np.expand_dims(np.sqrt(np.sum(kern**2,axis=-1)),-1)))
    kernx,kerny,kernz=kern.shape
    outim=np.zeros(testim.shape[:2])
    fillkern=kern.copy()
    denominator=length(kern.flatten())**2
    for xval in np.arange(0,testim.shape[0]-kernx-1,2):
        for yval in np.arange(0,testim.shape[1]-kerny-1,2):
            extract=testim[xval:xval+kernx,yval:yval+kerny,:kernz]
            if 'rot' in kernname:
                fillkern[kern == 1] = extract[kern == 1]
            outval=np.cos(angle(fillkern.flatten(),extract.flatten(),indenom=denominator))
            outim[xval,yval]=outval
    print(str(np.around(np.max(outim),3)) + ' max for ' + str(kernname))
    if np.max(outim) > .965:
        print('Detected! Localizing....')
        goodinds=np.where(outim == np.max(outim))
        goodrow,goodcol=goodinds
        goodrow,goodcol=goodrow[0],goodcol[0]
        boundbox=[goodrow, goodcol, goodrow+kernx, goodcol+kerny]
        detmvals.append(np.max(outim))
        detbbs.append(boundbox)
        print(boundbox)
badinds=[]
for bbind in range(len(detbbs)):
    if bbind in badinds:
        continue
    thisbb=np.array(detbbs[bbind])
    diffs=np.array([np.sum(abs(thisbb[:2] - x[:2])) for x in detbbs])
    if len(diffs[diffs < 16]) > 0:
        badinds.extend(np.where(diffs < 16)[0])
        badinds=[x for x in badinds if x != bbind]
for ele in sorted(badinds, reverse = True):  
    del detbbs[ele]
plt.imshow(testim)
print(detbbs)
for boundbox in detbbs:
    plt.scatter([boundbox[1],boundbox[3]],[boundbox[0],boundbox[2]],marker='x', color='red',s=15)
plt.title('RL-190, ' + str(len(detbbs)) + ' Total Detections')
plt.colorbar()
plt.show()
plt.close()
```

```python
testim=np.asarray(Image.open('data/RedLights2011_Medium/RL-195.jpg'))
testim=testim/np.max(testim)
for imname in glob('data/kernels/*.png'):
    kern=np.asarray(Image.open(imname))
    kern=kern/np.max(kern)
    kernx,kerny,kernz=kern.shape
    print(kernx,kerny)
    chanimlist=[]
    for chan in range(3):
        outimlist=[]
        for kcol in range(kern.shape[1]):
            kernstrip=kern[:,kcol,chan]
            convim=testim.copy()[:,:,chan]
            outim=np.zeros(convim.shape)
            for col in range(convim.shape[1]):
                imstrip=convim[:,col]
                conv=np.convolve(imstrip,kernstrip,mode='same')
                outim[:,col]=conv
            outimlist.append(outim)
        print(np.array(outimlist).shape)
        finalim=np.sum(np.array(outimlist),axis=0)
        chanimlist.append(finalim)
    finalfinalim=np.swapaxes(np.swapaxes(np.array(chanimlist),0,-1),0,1)
    finalfinalim=finalfinalim/np.max(finalfinalim)
    finalfinalim=np.sum(finalfinalim, axis=-1)
    plt.figure(figsize=(15,15))
    plt.imshow(finalfinalim)
    plt.colorbar()
    plt.show()
    plt.close()
```
